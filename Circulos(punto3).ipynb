```python
%pylab inline
from scipy.fftpack import fft, fftfreq
```

```python
#Se obtiene el archivo
data = np.array(loadtxt('BAO.dat'))
#Se pasa la matriz a su representacion compleja
dataComplex = np.array([[data[k][0] +1j*data[k][1]] for k in range(len(data))])
#grafica
scatter(np.real(dataComplex), np.imag(dataComplex), alpha = 0.071)
```

```python
#Transformada de fourier de la distribucion (funcion con imagen de variable compleja)
fftdat = fft(dataComplex)
#Espacio de frecuencias (discretas)
freq = fftfreq(len(fftdat), 1)
scatter(freq, np.abs(fftdat), alpha = 0.2)
```

Se puede ver que entre -0.3 y 0 hay una distribución uniforme de frecuencias con amplitudes aleatorias. Eso es lo que se conoce como ruido
El resto de la distribución es la que contiene las frecuencias no aleatorias: los círculos.
Si se omite ese intervalo de ruido, se puede limpiar la señal

```python
#Limpia el ruido
fftdat[((freq > -0.3) & (freq < 0))] = 0
#La distribución de puntos sin el ruido.
clean = ifft(fftdat)
scatter(np.real(clean), np.imag(clean), alpha = 0.4)
```

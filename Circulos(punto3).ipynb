# Punto 3: Círculos

## Teoría

La autoocorrelación de una función (definida como la convolución de la función con ella misma), puede ser interpretada como la similitud de la función con ella misma trasladada un vector r. En otras palabras, entre mayor sea el valor de la autocorrelación en r de una función, mayor similitud tendrá la función con ella misma trasladada dicha distancia $\vec{r}$.
Esto puede ser usado para determinar la frecuencia de las ocurrencias de máximos o mínimos periódicos en ciertas funciones no periódicas, dado que trasladada dicho periodo, la función será muy similar con ella misma.
$$$$

$$ C_{fg}(\vec{r}) = \int_{-\infty}^{\infty} f(\vec{x} - \vec{r}) \overline{g( \vec{x} )} d\vec{x} $$

$$$$

$$ C_{ff}(\vec{r}) = \int_{-\infty}^{\infty} f(\vec{x} - \vec{r}) \overline{f( \vec{x} )} d\vec{x} $$

$$$$

Esta interpretación se puede aplicar al problema de los círculos. Si se toma la correlación del conjunto de círculos, se debe esperar que tenga máximos espaciados por el diámetro de los círculos, ya que el conjunto de datos trasladado un vector con norma el diámetro del círculo es muy parecido al conjunto sin trasladar. Con esto, se espera que si se autocorrelacionan los datos, se pueda determinar el diámetro de los círculos con ruido.

En este orden de ideas, se puede moderlar la función de los círculos como una función de dos variables que es 1 en las coordenadas que aparecen en BAO.dat, y 0 en las que no. En estos términos, la función está compuesta de la función de círculos $s$ y la función de ruido $r$: 

$$$$

$$ f(\vec{x}) = s(\vec{x}) + r(\vec{x}) $$

$$$$

Si se saca la correlación de $ f $, se puede ver que es una suma de correlaciones y convoluciones de la siguiente manera.

$$$$

$$ C_{ff}(\vec{r}) = \int_{-\infty}^{\infty} (s(\vec{x} - \vec{r}) + r(\vec{x} - \vec{r}))(\overline{s(\vec{r}) + r(\vec{r})}) d\vec{x} $$

$$$$

$$ C_{ff}(\vec{r}) = \int_{-\infty}^{\infty} s(\vec{x} - \vec{r}) \overline{s( \vec{x} )} d\vec{x} + \int_{-\infty}^{\infty} s(\vec{x} - \vec{r}) \overline{r( \vec{x} )} d\vec{x}  +\int_{-\infty}^{\infty} r(\vec{x} - \vec{r}) \overline{s( \vec{x} )} d\vec{x} + \int_{-\infty}^{\infty} r(\vec{x} - \vec{r}) \overline{r( \vec{x} )} d\vec{x} $$

$$$$

$$ C_{ff}(\vec{r}) = C_{ss}(\vec{r}) + C_{sr}(\vec{r}) + C_{rs}(\vec{r}) + C_{rr}(\vec{r}) $$

$$$$

La correlación del ruido será claramente lineal, dado que por su homogeneidad, la función de ruido es muy parecida a ella misma trasladada con cualquier vector, sin embargo, como el conjunto es finito, cada vez habrá menor espacio sobre el que hallar la correlación ( el espacio de la interseccion los dominios de las funciones se reduce ). Esta autocorrelación se puede emular generando puntos aleatorios, y se puede demostrar que su transformada de fourier es casi insignificante al igual que la de las correlaciones mixtas respecto a la autocorrelación de los círculos solos. 

Esto significa que el ruido no afectará la periodicidad de los máximos de la autocorrelación de los círculos solos.
Así, podemos decir que la transformada de fourier de la autocorrelación de BAO.dat, es aproximadamente igual (en forma) a la transformada de fourier de los círculos solos en todos los puntos excepto en 0 (donde habrá una singularidad para la autocorrelación del ruido en 0 dado que no tiene un periodo dado ). Esto es:

$$$$

$$ F[C_{ff}](\vec{\nu}) \approx F[C_{ss}](\vec{\nu}) $$

$$$$

Ahora, usando el teorema de Wiener-Khinchin, se sabe que la transformada de fourier de una convolución es la multiplicación de las transformadas de las funciones, en particular, la transformada de fourier de una autocorrelación es la norma al cuadrado de la transformada de la función, esto es:

$$$$

$$ F[C_{ff}](\vec{\nu}) \approx F[C_{ss}](\vec{\nu}) = {\left\|{F[s](\vec{\nu})}\right\|}^2 $$

$$$$

De esta manera se puede determinar el diámetro de los círculos, analizando la transformada de fourier de la autocorrelación de BAO.dat. Sin embargo tanto la autocorrelación como la transformada son funciones vectoriales, así que para analizarlas mejor se promediarán sobre la norma de un vector (por anillos) y todo lo anterior se seguirá cumpliendo. Además de esto, para poder aplicar la transformada será necesario equiespaciar convertir BAO.dat a una matriz equiespaciada.


## Algoritmos necesarios

```python
# Método que devuelve una matriz simétrica equiespaciada a partir de un conjunto de datos en 2 dimensiones
# Cada punto se toma como si su funcion fuera f(x,y) = 1, y los puntos que no aparecen se toman como 0
# dat: el conjunto de datos
# mini: el minimo del intervalo
# maxi: el maximo del intervalo
# num: el numero 
# matrix: la matriz equiespaciada
def getSpacedMatrix( dati, num, mini, maxi):
    dat = dati.copy()
    # el espacio entre cada cuadro de la matriz
    dx = 1.0*(maxi - mini)/num
    # genera los linspaces que determinan la posición de inicio en x e y de cada columna y fila respectivamente
    linx = linspace(mini,maxi,num)
    liny = linspace(mini,maxi,num)
    # crea la matriz a retornar
    matrix = np.zeros((num,num))
    # Le asigna a cada cuadro el numero de puntos que caen dentro
    for i in range(len(dat)):
        iX = where(linx <= dat[i][0])[0][-1]
        iY = where(liny <= dat[i][1])[0][-1]
        matrix[iX][iY] += 1
    return matrix
```

```python
# Devuelve una lista con la matriz de abs²(fft2D) de una matriz equiespaciada dada, (en la primera posicion)
# y la matriz de frecuencias de dicha transformada (en la segunda posicion)
# matrix: la matriz a transformar con fourier
# dx: la el espacio entre los cuadros(el grosor de los cuadros) de la matriz a transformar
# fftmatrix: abs(fft(matrix)) matriz con valores reales
# freqMatrix: La matriz de frecuencias correspondiente (contiene la norma de la frecuencia en cada espacio)
# freqx: El vector de frecuencias los ejes de la matriz
def abs2fft( matrix2, dx ):
    # Copia la matriz para no afectarla
    matrix = matrix2.copy()
    # Saca a transformada de fourier 2D de la matriz
    Fm = fft2(matrix)
    # Genera los vectores de frecuencias en cada dimension y crea la matriz de frecuencias
    freqx = freqy = fftfreq(len(matrix), dx)
    freqMatrix  = np.zeros((len(freqx), len(freqy)))
    fftmatrix = np.zeros((len(freqx), len(freqy)))
    # Saca el valor absoluto de cada valor de la matriz
    i = j = 0
    for i in range(len(Fm)):
        for j in range(len(Fm[0])):
            fftmatrix[i][j] = float(abs(Fm[i][j]))
    #fftmatrix = Fm
    # Genera la matriz de frecuencias (con sus normas)
    i = j = 0
    for i in range(len(freqx)):
        for j in range(len(freqy)):
            freqMatrix[i][j] = (np.linalg.norm([freqx[i], freqy[j]]))
    
    return [fftmatrix, freqMatrix, freqx]
```

```python
# Genera un vector con el promedio del valor de la transformada de fourier en puntos donde la norma de la 
# frecuencia está en abs( fi + df), y su respectivo vector de frecuencias
# fftMatrix: la transformada de fourier (en abs) de una matriz
# freqx : el vector de frecuencias de los ejes de la matriz
# freqMatrix: la matriz de frecuencias respectiva de fftMatrix
# correlation: el promedio de la transformada(en valor absoluto) en un anillo de frecuencias
# freqxP : el valor de frecuencias para cada promedio
def meanFreqMatrix( fftmatrix, freqMatrix, freqx):
    
    # Genera el vector unidimensional de normas de frecuencias ( solo escoje las frecuencias en x)
    freqXP = freqx[where(freqx >= 0)]
    # Crea el vector unidimensional del promedio
    correlation = np.zeros(len(freqXP))
    # Recorre el vector de frecuencias y promedia sobre cada anillo
    for i in range(len(freqXP) - 1):
        # escoge los valores dentro del anillo
        index = where((freqMatrix > freqXP[i]) & (freqMatrix <= freqXP[i+1]))
        # promedia si encontró valores
        if(len(index[0]) != 0):
            vec = fftmatrix[index]
            correlation[i] = np.mean(vec)
    # El ultimo anillo        
    i = len(freqXP) -1
    index = where((freqMatrix > freqXP[i]))
    if(len(index[0]) != 0):
        vec = fftmatrix[index]
        correlation[i] = np.mean(vec)
    
    return [correlation, freqXP]
    
```

```python
# Devuelve un array con los primeros n maximos locales de una funcion  y = f(x) donde x e y son arrays
# vecX: los valores de x de la funcion
# vecy: los valores en y de la funcion
# n: el numero de maximos que se quiere hallar
# maxX : las posiciones de los maximos en x
# maxY : las posiciones de los maximos en y
def getMaxs( n, VecX, vecy): 
    # Copia el vector y para no afectarlo
    vecY = vecy.copy()
    # Crea los vectores de maximos
    maxX = []
    maxY = []
    # Busca los máximos
    m = 1
    i = 0
    while (i < len(VecX)) & ( m <= n ): 
        arg = i
        # Verifica que sean máximos locales
        if arg != 0 :
            if arg != len(VecX)-1 :
                if (vecY[arg - 1] <= vecY[arg]) & (vecY[arg + 1] <= vecY[arg]):
                    maxY.append( [vecY[arg]] )
                    maxX.append( [VecX[arg]] )
                    m += 1
            else: 
                if (vecY[arg - 1] <= vecY[arg]):
                    maxY.append( [vecY[arg]] )
                    maxX.append( [VecX[arg]] )
                    m +=1
        else: 
            if (vecY[arg + 1] <= vecY[arg]):
                maxY.append( [vecY[arg]] )
                maxX.append( [VecX[arg]] )
                m += 1
                
        i += 1
        
    # Convierte a array
    maxX = np.array(maxX)
    maxY = np.array(maxY)
    
    return np.array((maxX, maxY)).T
```

## Determinación del diámetro de los círculos

```python
# Se obtiene el archivo
data = np.array(loadtxt('BAO.dat'))
# Se emula el ruido
noise = 3 - 6*np.random.random((len(data), 2))

# grafica
scatter(data.T[0], data.T[1], alpha = 0.06)
#scatter(ruido.T[0], ruido.T[1])

# criterio para determinar el espaciado de la matriz a equiespaciar: un punto promedio por cuadro
string = "Se necesita una matriz de " + str( len(data) )
string += " cuadros aproximadamente\nEntonces su dimensión deberá ser de "+  str(np.sqrt(len(data)) )
string += "\nTómese dimensión = 200 por simplicidad"
print string
```

```python
# dimensión de la matriz equiespaciada
dim = 200

# Crea la matriz equiespaciada a partir de data
equimatrix = getSpacedMatrix(data,dim, -3, 3) 
# Crea la matriz equiespaciada a partir del ruido
equinoise = getSpacedMatrix(noise, dim, -3, 3)

# Muestra la matriz para verificar que es análoga al scatter de data
imshow(equimatrix)
#imshow(equinoise)
```

```python
# El espaciado de la matriz
dy = 6.0/dim

# Obtiene la matriz con las normas al cuadrado de la transformada de fourier de equimatrix
M = abs2fft( equimatrix, dy )
abs2FFT = M[0]
# Obtiene la matriz con las normas al cuadrado de la transformada de fourier de equinoise
Mnoise = abs2fft( equinoise, dy)
abs2FFTnoise = Mnoise[0]

# Obtiene la matriz de las frecuencias de BAO.dat
freqM = M[1]

# Obtiene el linspace de frecuencias de los ejes de BAO.dat
fx = M[2] 

# Obtiene el linspace de frecuencias de
# Grafica  ( las esquinas corresponden a los centros )
# por el orden en que el método fft2 y fftfreq presentan los datos, claramente la imagen correspondería 
# a una transdformada radialmente simétrica
imshow(np.log(abs(abs2FFT)))
```

```python
# Integra sobre anillos para promediar pasar la correlación de su forma vectorial a escalar (BAO.dat) y ruido
M1 = meanFreqMatrix( abs2FFT, freqM, fx )
M1noise = meanFreqMatrix( abs2FFTnoise, freqM, fx)

# La correlación como función escalar de los circulos y el ruido
Fcorrelat = M1[0]
FcorrelatNoise = M1noise[0]

# El correspondiente espacio de frecuencias escalares BAO.dat
fxP = M1[1]

# Grafica para verificar los máximos
#print len(fxP), len(Fcorrelat)

print "La singularidad en 0 puede ser explicada por la aleatoriedad de ciertos puntos"
plot(fxP, Fcorrelat )
```

```python
plot(fxP, FcorrelatNoise)
#plot(fxP, Fcorrelat)
plot(fxP, Fcorrelat - FcorrelatNoise)
```

Aqui se puede ver que la correlación de los puntos aleatorios es insignificante respecto a la correlación de los puntos con los círcuols, por lo que la aproximación propuesta en el marco teórico es válida

```python
# Selecciona los máximos de frecuencia
maxi = getMaxs(3, fxP, Fcorrelat)
#print maxi
# El diámetro de los círculos
diam = 1/maxi[0][1][0]
# El error asociado al cálculo
dd = dy
print diam, dd
```

El diámetro de los círculos es entonces el correspondiente al segundo máximo (el primero es causado por el ruido) de la transformada de fourier de la correlación de los datos, el cual es $ d = 1.5 \pm 0.06 $, donde el error es dado por dos veces la distancia del equiespaciado de la matriz.

Es importante resaltar que la precisión de la medición del diámetro del círculo está limitada por el tamaño de las dimensiones del área de distribucion de los círculos y puntos respecto al diámetro de los círculos, ya que la transfomada de fourier retorna solo frecuencias que son múltiplos enteros de $ 1/d_x $ $ $ o $ $ $1/d_y $, donde $d_x, d_y$ son las dimensiones del espacio de distribución. Así, se puede ver claramente que entre mayor sea la razon dimensiones-diámetro de círculo, mayor precisión se obtendrá. En este caso particular, el mínimo valor razonable que se podía obtener era $1.5$ sin importar el número de grillas que se usara en la matriz
